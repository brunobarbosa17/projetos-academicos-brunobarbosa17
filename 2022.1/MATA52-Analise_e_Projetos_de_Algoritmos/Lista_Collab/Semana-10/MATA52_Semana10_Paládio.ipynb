# 1. Mostre que um algoritmo que faz um número constante de chamadas a subrotinas de tempo polinomial tem também tempo polinomial.

Considerando um algoritmo que começa recebendo um input de tamanho $N$ e designando $S$ como a quantidade de subrotinas. Cada subrotina $S$ recebe o input, realiza seus passos e devolve um output. Logo, como cada subrotina roda em tempo polinomial o output também tem que ter tamanho polinomial proporcional ao tamanho do input. Logo, considerando $u$ o limite superior do grau do polinômio, é possível dizer que existe uma função:  
  
$p(N) = N^u + k$,  
  
tal que $p(N)$ é um limite superior para o tamanho do output de qualquer subrotina dado um input de tamanho $N$ e $k$ é uma constante qualquer.

Sendo $i$ uma variável de iteração, temos que:  
$u_{0} = u$  
$u_{i} = u_{i-1} + p(u_{i-1})$,  
para todo $1 <= i <= S$

Usando indução para provar que $u_{i}$ é polinomial em $u$, temos que $u_{i-1}$ é polinomial em $u$ (a composição de dois polinômios é em si um polinômio), chegamos a conclusão que $p(u_{i-1})$ é polinomial em $u$. A soma de dois polinomios também resulta em um polinômio, logo, $u_{i-1} + p(u_{i-1}) = u_{i}$ também é polinomial em $u$ e $u^S$ é um limite superior após a subrotina final, é também polinomial em $u$ consequentemente o tempo também deve ser polinomial.

# 3. Verdadeiro ou falso?  "Se X é um problema não polinomial então todas as instâncias de X são difíceis (ou seja, não podem ser resolvidas em tempo polinomial)." Prove com uma justificativa.

Falso. O diagrama de Venn proporciona uma visão sobre os diferentes tipos de problemas NP e sua relação entre si.

Screenshot from 2022-05-28 23-39-03.png

Considerando os problemas NP como um conjunto consistindo problemas que têm soluções em tempo polinomial, mas usando algoritmos não-determinísticos.

A região denominada P é o conjunto de problemas que existem algoritmos que podem resolvê-lo em tempo polinomial. Esta região é um subconjunto de problemas NP. Os problemas NP-Completo são a região de interseção dos problemas NP e os problemas NP-Difícil.

Problemas NP-Completo são aqueles problemas que têm uma solução em tempo polinomial, mas é derivado usando um algoritmo não-determinístico.

Se um problema NP-Difícil não possui um algoritmo não-determinístico, então ele não é NP-Completo, permanece NP-Difícil. Quando existe um algoritmo determinístico para o problema NP, ele é classificado como problema P. Quando existe um algoritmo não-determinístico para o problema, ele é classificado como NP-Completo. 

A questão generaliza a informação "Se X é um problema não polinomial então todas as instâncias de X são difíceis", ou seja, para ser verdadeiro teriamos que assumir que X é NP e toda instância de X são NP-Difícil. Porém, NP forma a classe base de todos os tipos de problemas, exceto alguns problemas NP-Difícil que não se enquadram na região NP.

Desse modo, não podemos afirmar que todas as instâncias de X serão NP-Difícil, podem existir instâncias que sejam NP-Completo ou P, e solucionadas em tempo polinomial.
